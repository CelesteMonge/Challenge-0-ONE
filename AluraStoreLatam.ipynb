#1. Análisis de facturación



```python

import pandas as pd


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular ingresos (suma de la columna Precio)
ingreso_1 = tienda["Precio"].sum()
ingreso_2 = tienda2["Precio"].sum()
ingreso_3 = tienda3["Precio"].sum()
ingreso_4 = tienda4["Precio"].sum()

# Crear un DataFrame resumen
ingresos = pd.DataFrame({
    "Tienda": ["1", "2", "3", "4"],
    "Ingreso Total": [ingreso_1, ingreso_2, ingreso_3, ingreso_4]
})

print(ingresos)
```

# 2. Ventas por categoría

```python
import pandas as pd


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Cantidad de productos vendidos por categoría en cada tienda

ventas_cat_1 = tienda.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_2 = tienda2.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_3 = tienda3.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_4 = tienda4.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)

print("Tienda 1 - Categorías más vendidas:")
print(ventas_cat_1.head())
print("\nTienda 2 - Categorías más vendidas:")
print(ventas_cat_2.head())
print("\nTienda 3 - Categorías más vendidas:")
print(ventas_cat_3.head())
print("\nTienda 4 - Categorías más vendidas:")
print(ventas_cat_4.head())
```

```python
# Distribución de categorías - Ejemplo Tienda 1
ventas_cat_1 = tienda.groupby("Categoría del Producto")["Producto"].count()
plt.figure(figsize=(6,6))
plt.pie(ventas_cat_1, labels=ventas_cat_1.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribución de ventas por categoría - Tienda 1")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio de cada tienda

promedio_1 = tienda["Calificación"].mean()
promedio_2 = tienda2["Calificación"].mean()
promedio_3 = tienda3["Calificación"].mean()
promedio_4 = tienda4["Calificación"].mean()

# Crear un DataFrame resumen
calificaciones = pd.DataFrame({
    "Tienda": ["1", "2", "3", "4"],
    "Calificación Promedio": [promedio_1, promedio_2, promedio_3, promedio_4]
})

print(calificaciones)
```

```python
# Relación Precio vs Reseña - Ejemplo Tienda 2
plt.figure(figsize=(6,4))
plt.scatter(tienda2["Precio"], tienda2["Calificación"], alpha=0.6, color="green")
plt.title("Relación entre Precio y Reseñas - Tienda 2")
plt.xlabel("Precio")
plt.ylabel("Reseña")
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# --- Tienda 1 ---
productos_1 = tienda.groupby("Producto")["Precio"].count().reset_index().sort_values(by="Precio", ascending=False)
print("Tienda 1 - Más vendidos:")
print(productos_1.head())
print("\nTienda 1 - Menos vendidos:")
print(productos_1.tail())

# --- Tienda 2 ---
productos_2 = tienda2.groupby("Producto")["Precio"].count().reset_index().sort_values(by="Precio", ascending=False)
print("\nTienda 2 - Más vendidos:")
print(productos_2.head())
print("\nTienda 2 - Menos vendidos:")
print(productos_2.tail())

# --- Tienda 3 ---
productos_3 = tienda3.groupby("Producto")["Precio"].count().reset_index().sort_values(by="Precio", ascending=False)
print("\nTienda 3 - Más vendidos:")
print(productos_3.head())
print("\nTienda 3 - Menos vendidos:")
print(productos_3.tail())

# --- Tienda 4 ---
productos_4 = tienda4.groupby("Producto")["Precio"].count().reset_index().sort_values(by="Precio", ascending=False)
print("\nTienda 4 - Más vendidos:")
print(productos_4.head())
print("\nTienda 4 - Menos vendidos:")
print(productos_4.tail())
```

```python
# Función para graficar los top 5 más vendidos
def graficar_top(df, tienda_nombre):
    top5 = df.head(5)
    plt.figure(figsize=(8,4))
    plt.bar(top5["Producto"], top5["Precio"], color="skyblue")
    plt.title(f"Top 5 productos más vendidos - Tienda {tienda_nombre}")
    plt.xticks(rotation=45)
    plt.ylabel("Cantidad vendida")
    plt.show()

# Graficar para cada tienda
graficar_top(productos_1, "1")
graficar_top(productos_2, "2")
graficar_top(productos_3, "3")
graficar_top(productos_4, "4")
```

## 1 Introducción
El presente informe tiene como objetivo **ayudar al Sr. Juan a decidir qué tienda de su cadena Alura Store debería vender** para iniciar un nuevo emprendimiento. Para ello, se analizaron datos de **ventas, ingresos, reseñas, productos y costos de envío** de las cuatro tiendas.  
El análisis incluye **visualizaciones gráficas** que permiten identificar tendencias, fortalezas y debilidades de cada tienda.

## 2️ Desarrollo

### 2.1 Ingresos Totales por Tienda
Los ingresos totales se calcularon sumando los precios de todos los productos vendidos:  

- Tienda 1: 1.150.880.000  
- Tienda 2: 1.116.344.000  
- Tienda 3: 1.098.020.000  
- Tienda 4: 1.038.376.000  

**Insight:** La tienda 4 tiene los ingresos más bajos, mientras que la tienda 1 genera los ingresos más altos.

---

### 2.2 Categorías de Productos más Vendidas
Se analizó la cantidad de productos vendidos por categoría en cada tienda:  

- **Tienda 1:** Muebles (465), Electrónicos (448), Juguetes (324)  
- **Tienda 2:** Muebles (442), Electrónicos (422), Juguetes (313)  
- **Tienda 3:** Muebles (499), Electrónicos (451), Juguetes (315)  
- **Tienda 4:** Muebles (480), Electrónicos (451), Juguetes (338)  

**Insight:** Todas las tiendas venden principalmente Muebles y Electrónicos. La tienda 4 tiene menor número de ventas en categorías clave como Electrodomésticos y Deportes, indicando menor diversidad y demanda.

---

### 2.3 Calificación Promedio de Clientes
El promedio de reseñas indica la satisfacción del cliente:  

- Tienda 1: 3.98  
- Tienda 2: 4.04  
- Tienda 3: 4.05  
- Tienda 4: 4.00  

**Insight:** La tienda 4 no destaca en satisfacción del cliente; aunque no es la peor, está por debajo de las tiendas 2 y 3.

---

### 2.4 Productos Más y Menos Vendidos
**Tienda 4:**  

- Más vendidos: Cama box, Cubertería, Cama king, Dashboards con Power BI, Carrito de control remoto  
- Menos vendidos: Ciencia de datos con Python, Refrigerador, Guitarra acústica, Armario, Guitarra eléctrica  

**Insight:** Algunos productos importantes en demanda tienen ventas bajas, afectando la eficiencia general de la tienda.

---

### 2.5 Costo de Envío Promedio
- Tienda 1: 26.018  
- Tienda 2: 25.216  
- Tienda 3: 24.806  
- Tienda 4: 23.459  

**Insight:** Aunque la tienda 4 tiene el costo de envío más bajo, esto no compensa sus ingresos más bajos y menor diversidad de ventas.

---

## 3️ Conclusión y Recomendación
Tras analizar todos los factores:  

- **Ingresos totales:** La tienda 4 es la que genera menos ingresos.  
- **Distribución de categorías:** Menor diversidad y ventas en categorías clave.  
- **Calificación promedio de clientes:** Leve desventaja frente a tiendas 2 y 3.  
- **Productos más y menos vendidos:** Algunos productos de interés tienen ventas bajas.  
- **Costo de envío promedio:** Es el más bajo, pero no suficiente para compensar otras debilidades.  

**Recomendación:** Se sugiere que el Sr. Juan **venda la tienda 4**, ya que presenta la menor eficiencia en términos de ingresos y ventas diversificadas. Mantener las tiendas 1, 2 y 3 permitirá enfocar recursos en unidades con mayor potencial de crecimiento y clientes más satisfechos.
